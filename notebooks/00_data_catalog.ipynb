```python
# IMPORTANDO FUNÇÕES PARA ENCONTRAR A PASTA DO DIRETÓRIO ROOT (que contém a src)
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
# IMPORTANDO MÉTODOS DA CAMADA SILVER
from src.silver.reader import read_bronze_table
```

```python
# DEFININDO TABELAS QUE VAMOS LER

TABLES = [
    'alunos',
    'municipio',
    'uf',
    'meta_municipio',
    'meta_uf',
    'meta_brasil'
]
```

```python
# FAZENDO LEITURA DAS TABELAS COM src.silver.read_bronze_table
import pandas as pd

catalog_rows = []

for table_name in TABLES:
    df = read_bronze_table(table_name)

    catalog_rows.append({
        'layer': 'bronze',
        'table_name': table_name,
        'rows': len(df),
        'columns': len(df.columns),
        'column_names': list(df.columns),
    })

catalog = pd.DataFrame(catalog_rows)
catalog
```

```python
# ADICIONANDO AS CHAVES AO CATÁLOGO

PRIMARY_KEYS = {
    'alunos': ['ano', 'id_aluno'],
    'municipio': ['ano', 'id_municipio', 'rede'],
    'uf': ['ano' ,'sigla_uf', 'rede'],
    'meta_municipio': ['ano', 'id_municipio'],
    'meta_uf': ['ano', 'sigla_uf'],
    'meta_brasil': ['ano'],
}

catalog['primary_key'] = catalog['table_name'].map(PRIMARY_KEYS)

catalog
```

```python
# ADICIONANDO DESCRIÇÃO DAS TABELAS
TABLE_DESCRIPTIONS = {
    'alunos': 'Dados individuais dos alunos participantes da avaliação.',
    'municipio': 'Indicadores agregados por município.',
    'uf': 'Indicadores agregados por unidade da federação.',
    'meta_municipio': 'Metas de alfabetização por município.',
    'meta_uf': 'Metas de alfabetização por unidade da federação.',
    'meta_brasil': 'Metas nacionais de alfabetização.'
}

catalog['description'] = catalog['table_name'].map(TABLE_DESCRIPTIONS)

catalog
```
